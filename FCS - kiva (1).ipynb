### 1. Normalize the loan_lenders table. In the normalized table, each row must have one loan_id and one lender.

```python
loanslenders.head()
```

### 12. For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

```python
# selezioniamo le colonne necessario all'analisi e sistemiamo il formato dalla colonna duration
loans_utili = loans[['loan_id', 'disburse_time', 'planned_expiration_time', 'loan_amount', 'duration']]
loans_utili['duration'] = loans_utili['duration'].dt.days

loans_utili.head()
```

```python
# eliminaimo il fuso orario
loans_utili["disburse_time"].dt.tz_localize(None)
loans_utili["planned_expiration_time"].dt.tz_localize(None)
```

```python
# selezioniamo ed eliminiamo i dati mancanti
loans_utili_no_missing = loans_utili.dropna()
loans_utili_no_missing.isnull().sum()
```

```python
# la funzione prende le date di inizio e fine, e ritorna il totale dei giorni che intercorrono tra queste due date, 
# dividendo per anno
def n_days (date1, date2):
    (begin,end) = (date1,date2) if date1<date2 else (date2,date1)
    date_list = list()
    date_list.append(begin)
    for y in range(begin.year+1, end.year+1):
        date_list.append(pd.Timestamp(year=y, month=1, day=1, tz='utc')) 
    date_list.append(end) 
    
    p12 = dict()
    for i in range(len(date_list)-1):
        p12[date_list[i].year] = (date_list[i+1] - date_list[i]).days 

    return pd.Series(p12, name='DateValue')
```

```python
days = loans_utili_no_missing.apply(lambda x: n_days(x['planned_expiration_time'],x['disburse_time']), axis = 1)
days.head()
```

```python
days.fillna(0, inplace= True)
days = pd.concat([days, loans_utili_no_missing['loan_id']
                 , loans_utili_no_missing['loan_amount']
                 , loans_utili_no_missing['duration'] ]
                , axis = 1, sort = True)

days.head()
```

```python
# creiamo la tabella amount dove per ogni loan id dividiamo i giorni trascorsi ad ogni anno per il numero di giorni totali (duration)
# e moltiplichiamo questo valore per la sonna totale di denaro (loan_amount)
amounts = days[[2011,2012,2013,2014,2015,2016,2017,2018]].div(days.duration, axis=0).mul(days.loan_amount, axis = 0)
 
# riconcateniamo loan_id e rinnomianiamo le colonne e ribaltiamo gli assi 
amounts = pd.concat([days['loan_id'], amounts],  axis = 1 ) 
amounts = amounts.melt(id_vars=["loan_id"],
        var_name="Year", 
        value_name="amount_per_year")
 
# raggruppiamo per anno
amounts.groupby(['Year']).sum()
```

Nella tabella sopra vediamo per ogni anno quanti soldi sono stati chiesti in prestito (considerando le proporzioni dei giorni effettivi in ogni anno).
